```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df1 = pd.read_csv("data1.txt", sep = " ", header = None, names = ["x", "right", "central"])
x = df1["x"].to_numpy()
right = df1["right"].to_numpy()
central = df1["central"].to_numpy()

plt.plot(x, right, label = "right(1)")
plt.plot(x, central, label = "central(2)")
plt.legend()
plt.show()
```

```python
df2 = pd.read_csv("data2.txt", sep = " ", header = None, names = ["x", "central22"])
x = df2["x"].to_numpy()
central22 = df2["central22"].to_numpy()

df3 = pd.read_csv("data3.txt", sep = " ", header = None, names = ["x", "central24"])
central24 = df3["central24"].to_numpy()

plt.plot(x, central22, label = "second derivative(2)")
plt.plot(x, central24, label = "second derivative(4)")
plt.legend()
plt.show()
```
