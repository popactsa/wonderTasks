```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return (x/10)**np.sin(x)


x = []
y = []
fun_values = []
with open('data_numerical_interpolate.txt', 'r') as f:
    data = ['1']
    while (data != []):
        data = list(f.readline().split())
        if data == []:
            continue
        print(data)
        x_temp, y_temp = float(data[0]), float(data[1])
        x.append(x_temp)
        y.append(y_temp)
x_fun = np.linspace(0, 10, 201)
for i in x_fun:
    fun_values.append(func(i))
plt.plot(x, y)
plt.plot(x_fun, fun_values, '*')
plt.ylim(0, 3)
plt.show()
```
